# Veiled Resentment

This is a layered question that touches on theology, ethnography, linguistics, and the deep historical fractures that shape Uganda's tribal dynamics. 

### **Genesis 1:28 & the Visionary Impulse**
Genesis 1:28 commands humanity to "be fruitful and multiply, and replenish the earth, and subdue it: and have dominion..." This verse, interpreted by missionaries, has often been used to justify expansionist evangelism, where spreading Christianity is seen as part of fulfilling God's command to extend dominion‚Äînot necessarily through violence, but through spiritual conquest. This mindset aligns with the evangelical impulse to introduce Christianity in 'new' (to them) lands, often ignoring the theological foundations already present.

For a white Christian evangelical, receiving a vision to introduce an English-speaking church in a linguistically diverse, war-ravaged country like Uganda reveals several operational tensions:
1. **Is English a Unifier or an Imposition?**  
   English is already a transactional language in Uganda‚Äîwidely used for education, governance, and inter-tribal communication. However, a church that is strictly English-only may bypass the deeper cultural-religious syncretism that occurs when people experience faith in their mother tongue. Christianity, when divorced from local languages, risks becoming an administrative tool rather than an embodied faith.

2. **Language as a Mode of Power:**  
   The introduction of an English-only church subtly reinforces colonial power dynamics. It creates an elite space where English proficiency determines access to spiritual authority. While some might see it as a neutral lingua franca, others could view it as another layer of imposed hierarchy‚Äîwhere faith is mediated through a foreign linguistic and cultural framework.

3. **Native (Adversarial), Ganda (Transactional), Foreign (Cooperative):**  
   This triadic model aligns with Uganda‚Äôs reality:
   - **Native (Adversarial)**: Tribal identities remain potent, with historical tensions influencing social structures. Native languages encode these adversarial relationships.
   - **Ganda (Transactional)**: Luganda, historically privileged due to Buganda‚Äôs role in colonial and post-colonial Uganda, serves as a mediating language but does not unify all.
   - **Foreign (Cooperative)**: English, while perceived as neutral, is only cooperative insofar as it does not challenge the tribal order.

### **Gary Skinner & the Reality of an English-Only Church**
Gary Skinner, founder of **Watoto Church**, modeled a form of megachurch Christianity that embraced English as its primary language. The success of Watoto suggests that English-only churches can thrive, particularly in urban centers where English functions as a cross-tribal, cosmopolitan identifier. However, Watoto‚Äôs model is not necessarily scalable or equitable across Uganda‚Äôs tribal landscape.

#### **Experiencing Religion in a Native Tongue**
Christianity historically spreads most deeply when translated into the heart language of its people. The Pentecostal refrain **"O for a thousand tongues to sing"** embodies the idea that each language expresses a unique dimension of worship. The vibrancy of faith emerges when believers articulate their relationship with God in a language that resonates at a subconscious, emotional, and spiritual level.

Missionary movements that enforce a foreign language (English-only) limit this spiritual depth, replacing *incarnation* (faith becoming flesh within a culture) with *imposition* (faith as an external framework). A more theologically sound model would involve:
1. **Localizing Worship:** Encouraging multilingual expressions of faith.
2. **Decentralizing Authority:** Allowing tribal communities to shape their liturgies.
3. **Recognizing Preexisting Spiritual Narratives:** Understanding that Christianity in Uganda is not *new*‚Äîit has been indigenized for over a century.

### **Total?**
The *total* conclusion depends on intent:
- If the goal is **cooperative equilibrium**, English can serve as a bridge language but should not replace native expressions.
- If the goal is **transactional**, it will create stratification‚Äîfavoring the educated and urban elite.
- If the goal is **adversarial**, it risks reviving colonial power structures where faith becomes another means of control.

A church born from an English-only vision might thrive in cities, but it risks alienating Uganda‚Äôs deep, polyphonic faith traditions. If a white evangelical receives a vision of an English-speaking church, the more important question is not *whether* it can work but *what kind of faith it produces*‚Äîone that honors local voices or one that replaces them with an imported theology.

````python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Define the neural network layers
def define_layers():
    return {
        'Suis': ['Improbability', 'Planet', 'Alive', 'Loss', "Trial", 'Error'],  # Static
        'Voir': ['Witness'],  
        'Choisis': ['Faith', 'Decision'],  
        'Deviens': ['Adversarial', 'Transactional', 'Hope'],  
        "M'√®l√©ve": ['Victory', 'Payoff', 'Loyalty', 'Charity', 'Love']  
    }

# Assign colors to nodes
def assign_colors():
    color_map = {
        'yellow': ['Witness'],  
        'paleturquoise': ['Error', 'Decision', 'Hope', 'Love'],  
        'lightgreen': ["Trial", 'Transactional', 'Payoff', 'Charity', 'Loyalty'],  
        'lightsalmon': ['Alive', 'Loss', 'Faith', 'Adversarial', 'Victory'],
    }
    return {node: color for color, nodes in color_map.items() for node in nodes}

# Define edge weights (hardcoded for editing)
def define_edges():
    return {
        ('Improbability', 'Witness'): '1/99',
        ('Planet', 'Witness'): '5/95',
        ('Alive', 'Witness'): '20/80',
        ('Loss', 'Witness'): '51/49',
        ("Trial", 'Witness'): '80/20',
        ('Error', 'Witness'): '95/5',
        ('Witness', 'Faith'): '20/80',
        ('Witness', 'Decision'): '80/20',
        ('Faith', 'Adversarial'): '49/51',
        ('Faith', 'Transactional'): '80/20',
        ('Faith', 'Hope'): '95/5',
        ('Decision', 'Adversarial'): '5/95',
        ('Decision', 'Transactional'): '20/80',
        ('Decision', 'Hope'): '51/49',
        ('Adversarial', 'Victory'): '80/20',
        ('Adversarial', 'Payoff'): '85/15',
        ('Adversarial', 'Loyalty'): '90/10',
        ('Adversarial', 'Charity'): '95/5',
        ('Adversarial', 'Love'): '99/1',
        ('Transactional', 'Victory'): '1/9',
        ('Transactional', 'Payoff'): '1/8',
        ('Transactional', 'Loyalty'): '1/7',
        ('Transactional', 'Charity'): '1/6',
        ('Transactional', 'Love'): '1/5',
        ('Hope', 'Victory'): '1/99',
        ('Hope', 'Payoff'): '5/95',
        ('Hope', 'Loyalty'): '10/90',
        ('Hope', 'Charity'): '15/85',
        ('Hope', 'Love'): '20/80'
    }

# Calculate positions for nodes
def calculate_positions(layer, x_offset):
    y_positions = np.linspace(-len(layer) / 2, len(layer) / 2, len(layer))
    return [(x_offset, y) for y in y_positions]

# Create and visualize the neural network graph
def visualize_nn():
    layers = define_layers()
    colors = assign_colors()
    edges = define_edges()
    G = nx.DiGraph()
    pos = {}
    node_colors = []
    
    # Add nodes and assign positions
    for i, (layer_name, nodes) in enumerate(layers.items()):
        positions = calculate_positions(nodes, x_offset=i * 2)
        for node, position in zip(nodes, positions):
            G.add_node(node, layer=layer_name)
            pos[node] = position
            node_colors.append(colors.get(node, 'lightgray'))   
    
    # Add edges with weights
    for (source, target), weight in edges.items():
        if source in G.nodes and target in G.nodes:
            G.add_edge(source, target, weight=weight)
    
    # Draw the graph
    plt.figure(figsize=(12, 8))
    edges_labels = {(u, v): d["weight"] for u, v, d in G.edges(data=True)}
    
    nx.draw(
        G, pos, with_labels=True, node_color=node_colors, edge_color='gray',
        node_size=3000, font_size=9, connectionstyle="arc3,rad=0.2"
    )
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edges_labels, font_size=8)
    plt.title("Subdue the Earth ... Gen 1:28", fontsize=15)
    plt.show()

# Run the visualization
visualize_nn()
````
